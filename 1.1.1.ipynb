```python
%matplotlib widget

from sympy.abc import U

from plot import plot_g, plot_r, plot_i
```

```python
a0 = 5 * 10 ** -4
a1 = 4
Ulim = (-2, 2)
```

```python
# 1
G = a0 * (a1 - U)
plot_g(G, xlim=Ulim, ylim=(0, 0.005))
```

```python
R = 1 / G
plot_r(R, xlim=Ulim, ylim=(0, 1000))
```

```python
# 2
I = U * G
plot_i(I, xlim=Ulim, ylim=(-0.007, 0.003))
```

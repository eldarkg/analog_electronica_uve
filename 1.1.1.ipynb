```python
%matplotlib widget

from sympy.abc import U

from visual import plot
```

```python
a0 = 5E-4
a1 = 4
Ulim = (-2, 2)
```

```python
# 1
G = a0 * (a1 - U)
plot(G, xlim=Ulim, ylim=(0, 0.005))
```

```python
R = 1 / G
plot(R, xlim=Ulim, ylim=(0, 1000))
```

```python
# 2
I = U * G
plot(I, xlim=Ulim, ylim=(-0.007, 0.003))
```
